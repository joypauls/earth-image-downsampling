```python
from PIL import Image
```

```python
# plt.imshow(img)
```

```python
img = Image.open("earth.png")

img
```

```python
compressed_size = (26, 26)
img_compressed = img.resize(compressed_size, resample=Image.Resampling.BILINEAR)

img_compressed
```

```python
img_restored = img_compressed.resize(img.size, Image.Resampling.NEAREST)

img_restored
```

```python
img_restored.save("earth_downsampled.png")
```
